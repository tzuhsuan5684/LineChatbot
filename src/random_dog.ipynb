```python
import requests
import json
import webbrowser
import time
from IPython.display import Image, display

# 設定 API 的網址
url = "https://dog.ceo/api/breeds/image/random"

print("🐶 正在聯絡狗狗總部 (API)...")

# 1. 發送請求 (Request)
# 這就像是透過電話向餐廳點餐
response = requests.get(url)

# 2. 檢查狀態 (Status Code)
# 200 代表成功，404 代表找不到，500 代表伺服器壞了
if response.status_code == 200:
    print("✅ 連線成功！正在解析資料...")

    # 3. 解析 JSON 資料
    # 將回傳的文字轉換成 Python 的字典 (Dictionary)
    data = response.json()

    # 顯示完整的資料結構
    print("\n--- API 回傳的完整內容 (JSON) ---")
    print(json.dumps(data, indent=4)) # indent=4 讓排版漂亮一點
    print("-------------------------------\n")

    # 4. 提取我們要的資訊
    # 我們只需要 'message' 欄位裡的圖片網址
    image_url = data['message']

    print(f"🎉 找到狗狗了！網址是：\n{image_url}")

    # 5. 實際打開圖片 (視覺化回饋)
    print("\n📸 圖片顯示如下：")
    display(Image(url=image_url, width=400))

else:
    print(f"❌ 發生錯誤，狀態碼：{response.status_code}")
```
